```python
import csv, pandas
```

```python
namesa = ["variablename", "fieldlength"]
buf = pandas.read_csv('psu031alay_extraced.csv', names = namesa)
```

```python
a = list(buf.variablename)
b = list(buf.fieldlength)
a = a[1:]
b = b[1:]
for index in range(len(a)):
    a[index] = a[index].rstrip()
for index in range(len(b)):
    if b[index].find('*') != -1:
        b[index] = b[index][:b[index].find('*')]
    b[index] = int(b[index].rstrip())    
print(a)
print(b)
        
```

```python
buf = pandas.read_fwf('mybox-selected/sc031aow.txt', index_col = False, \
                      widths = b, header = None, names = a)
buf
```
