```python
import pandas as pd
```

```python
#read the csv that we want to concatenate
asc_uni = pd.read_csv("ACS_2016_school-district-unified_FULL.csv", header = 1, encoding = "latin1")
asc_ele = pd.read_csv("ACS_2016_school-district-elementary_FULL.csv", header = 1, encoding = "latin1")
asc_sec = pd.read_csv("ACS_2016_school-district-secondary_FULL.csv", header = 1, encoding = "latin1")
```

```python
#turn their unique identifier to a common type
asc_uni['Geo_FIPS'] = asc_uni.Geo_FIPS.astype('float64')
asc_ele['Geo_FIPS'] = asc_ele.Geo_FIPS.astype('float64')
asc_sec['Geo_FIPS'] = asc_sec.Geo_FIPS.astype('float64')
```

```python
#concate them using pandas.concat
#input is a list of dataframe that you want to concatenate
concated = pd.concat([asc_uni, asc_ele, asc_sec])
```

```python
#turn the concatenated file to csv
concated.to_csv("acs_concat.csv", index = False)
```
